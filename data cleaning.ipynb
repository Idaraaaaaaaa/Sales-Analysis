```python
new_columns=df.select_dtypes(include=['number'])
sns.heatmap(new_columns.corr(),cmap='PuBuGn',annot=True)
plt.show()
```

```python
df_Region=df['Region'].value_counts().reset_index().head(5)
df_Region.rename(columns={'index': 'Region','Region':'Count'}, inplace=True)
df_Region
```

```python
sns.barplot(data=df_Region,x='Region',y='Count',alpha=0.7,color='blue')
plt.xlabel('Region')
plt.ylabel('count')
plt.title('Region vs Count Plot')
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize=(60, 14))
sns.catplot(data=df,x='Region',y='Units',kind='bar')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(14, 8))

sns.scatterplot(data=df,x='Region',y='Units')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(14, 8))
df.Region.value_counts().plot(kind='pie', autopct="%0.1f%%", explode=[0,0,0], shadow=True)
plt.show()
```
